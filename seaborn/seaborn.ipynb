## JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart displays the correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot 

### Univariate Analysis

```python
sns.jointplot(x= 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

## Pair plot
A "pairs plot" is also known as scatterplot, in which one variable in the same data row is matched with another variables value, like this : Pairs plots are just elaborations on this showing all variables paired with all the other variables 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

## Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde = False, bins = 10)
```

## Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about 

- box plot
- violinplot
- coutplot
- bar plot

```python
## Count plot

## ON review
sns.countplot(data = df)
```

```python
df_for_countplot = df.drop(columns=['total_bill', 'tip', 'size'])
df_for_countplot
```

```python
sns.countplot(x = 'sex', data = df)
```

```python
sns.countplot(y = 'sex', data = df)
```

```python
## Bar plot 
sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```python
# Count plot

sns.countplot(x = 'day', data = df)
```

```python
# Bar plot 

sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```python
## Bar plot 
sns.barplot(x = 'sex', y = 'total_bill', data = df)
```

## Box plot 
A box and whisker plot (sometimes call a boxplot) is a graph that presents information from a five number summarty

```python
## Box plot 
sns.boxplot(x = 'sex', y = 'total_bill', data = df)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df, palette = 'rainbow')
```

```python
sns.boxplot(data = df, orient = 'v')
```

```python
# Categorize my data based on some other categories

sns.boxplot(x = 'total_bill', y = 'day', hue = 'smoker', data = df)
```

## Violin plot
Violin plot helps us to see both the distribution of data in terms of kernel density estimation and box plot

```python
sns.violinplot(x= 'total_bill', y = 'day', data = df, palette = 'rainbow')
```
