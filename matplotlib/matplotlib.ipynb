# Matplotlib

```python
import matplotlib.pyplot as plt

import numpy as np

%matplotlib inline
```

```python
x  = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
# plotting using matplotlib
# plt scatter 

plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')

# If you want to save your figure
plt.savefig('Test.png')
```

```python
y = x*x
```

```python
## plt plot

plt.plot(x,y,'r*', linestyle = 'dashed', linewidth = 2, markersize = 12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d Diagram')
```

```python
# Creating subplots 

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y*--')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
```

```python
x = np.arange(0,4*np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')


plt.plot(x,y)
```

```python
x = np.arange(0,5*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)


#subplotting 

plt.subplot(2,1,1)
plt.plot(x,y1, 'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sine wave graph')



plt.subplot(2,1,2)
plt.plot(x,y2, 'b*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('cosine wave graph')
```

## Bar Plot

```python
## Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15, 7]
plt.bar(x,y, align = 'center')
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

## Histograms

```python
a = np.array([22,87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
plt.hist(a)
plt.title('histogram')
plt.show()
```

## Box plot using matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data, vert= True, patch_artist = True);
```

## Pie chart

```python
labels = 'python', 'c++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral','lightskyblue']
explode = (0.1, 0,0,0)

plt.pie(sizes, explode = explode , labels = labels, colors = colors,
       autopct = '%1.1f%%', shadow = True)

plt.axis('equal')
plt.show()
```
