## CSV

```python
from io import StringIO, BytesIO
```

```python
data = ('col1, col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3'
       )
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
df = pd.read_csv(StringIO(data), usecols = ['col1', 'col3'])
```

```python
df
```

```python
df.to_csv('Test.csv')
```

```python
data = ('a,b,c,d\n'
       '1,2,3,4\n'
       '5,6,7,8\n'
       '9,10,11')
```
